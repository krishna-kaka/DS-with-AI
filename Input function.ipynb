# Input()

- Input function in Python is used to take input from the user. When this function is called, the program waits for the user to type something and press Enter. 
- The input received is always in the form of a string.
- For the interpreter to convert the input into other data types like integer or float, explicit type conversion is required using functions like int() or float().
- Example:
    - name = input("Enter your name: ")     # Taking string input which is default
    - age = int(input("Enter your age: "))  # Converting input to integer
    - height = float(input("Enter your height in meters: "))    # Converting input to float

- Slicing can be applied to the string received from input to extract specific parts of the string.


```python
x = input() # Input function to take input from user
y = input()

z = x + y
print(z)
```

```python
x1 = input('Enter the first number : ')
y1 = input('Enter the second number : ')

z1 = x1 + y1

print(z1)
```

```python
print(type(x1)) # To check the data type of input variable
print(type(y1)) # To check the data type of input variable
print(type(z1)) # To check the data type of output variable
```

```python
x2=int(input('Enter First number:'))    # Type casting to convert string to integer
y2=int(input('Enter Second Number:'))   # Type casting to convert string to integer, here the input parsed will be integer

z2=x2+y2
print(z2)
```

```python
print(type(x2)) # To check the data type of input variable after type casting
print(type(y2)) # To check the data type of input variable after type casting
print(type(z2)) # To check the data type of input variable after type casting
```

```python
x3 = input('Username : ')
y3 = input('Password : ')

z3 = x3 + y3
print(z3)
```

```python
st = input('enter a string : ')
print(st)
```

```python
print(st[0])
```

```python
st = input('enter a string : ')[1]  # indexing in input function, prints the value at index 1 to the output
print(st)
```

```python
st = input('enter a string : ')[3]  # prints the value at index 3 to the output 
print(st)
```

```python
st = input('enter a string: ')[3:8] # Slicing in input : prints the value from index 3 to 7 to the output
print(st)
```

```python
result=input('Enter an expression :')   # input function to take an expression from user, what ever input expression is givent he same is printed as output as the default input type is string
print(result)
print(type(result))
```

```python
result=int(input('enter an expression : ')) # The statement gives error as we are trying to parse a mathematical expression as integer directly
print(result)
```

```python
# eval function is used to evaluate the expression present in the string format and gives the output
# the function then converts the string expression into a valid expression and then evaluates it to give the output
result=eval(input('Enter an expression :'))  
print(result)
```
