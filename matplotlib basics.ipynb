# MatplotLib

- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization in various fields such as data science, machine learning, and scientific research. Matplotlib provides a flexible and powerful way to create a wide range of plots and charts, from simple line graphs to complex 3D visualizations.
- Some key features of Matplotlib include:
    1. Versatility: Matplotlib can create a variety of plot types, including line plots, scatter plots, bar charts, histograms, pie charts, and more.
    2. Customization: Matplotlib offers extensive customization options, allowing users to modify every aspect of a plot, including colors, labels, titles, axes, and legends.
    3. Integration: Matplotlib can be easily integrated with other libraries such as NumPy, Pandas, and SciPy, making it a powerful tool for data analysis and visualization.
    4. Interactive Plots: Matplotlib supports interactive plots that can be embedded in web applications using libraries like Flask or Django, or displayed in Jupyter Notebooks.
    5. Publication Quality: Matplotlib can produce high-quality plots suitable for publication in scientific journals and presentations.

Boxplot

- Boxplot function allows us to visualize the distribution of data based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
- The red line in the boxplot represents the median of the data.
- The box represents the interquartile range (IQR) which contains the middle 50% of the data.
- The whiskers extend to the minimum and maximum values within 1.5 * IQR from the quartiles.
- Any data points outside of this range are considered outliers and are plotted as individual points.
- The boxplot provides a visual summary of the data distribution, making it easy to identify potential outliers and the overall spread of the data.
- The upper whisker extends to the largest data point within 1.5 times the interquartile range (IQR) above the third quartile (Q3).
- The lower whisker extends to the smallest data point within 1.5 times the IQR below the first quartile (Q1).


```python
data3 = np.random.randn(100)    # generate 100 random numbers from standard normal distribution
print(data3)

plt.boxplot(data3)  # plotting data3 as box plot

plt.show();
```

Contour Plot

- Contour plot is used to represent three dimensional data in two dimensions using contour lines
- Contour lines connect points of equal value
- The contour plot is useful for visualizing the shape of the data
- It can reveal patterns and relationships in the data
- The denser the contour lines, the steeper the slope they represent
- Contour plots are commonly used in fields such as meteorology, geology, and engineering
- the gradient of the function is perpendicular to the contour lines

```python
matrix1 = np.random.rand(10, 20)    # generate a 10x20 matrix with random numbers

cp = plt.contour(matrix1)   # plotting contour plot of matrix1

plt.show()
```

Handling X and Y ticks

- There are multiple ways to set axis limits in matplotlib:
 1. Using plt.axis([xmin, xmax, ymin, ymax]) function
 2. Using plt.xlim([xmin, xmax]) and plt.ylim([ymin, ymax]) functions
 3. using xticks() and yticks() functions
 The arguments for these functions are lists containing the limits for respective axes

```python
u = [5, 4, 9, 7, 8, 9, 6, 5, 7, 8]  # sample data for plotting

plt.plot(u)

plt.xticks([2, 4, 6, 8, 10])   # set x axis ticks at specified positions
plt.yticks([2, 4, 6, 8, 10])   # set y axis ticks at specified positions

plt.show()
```

Legend

- Legends are used to label each data series in the plot for better understanding
- They help in distinguishing between multiple plots
- Legend is added using the legend() function

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['Normal','Fast','Slow']);

# This method of adding legend is prone to errors if plots are added or removed. 
# If the curves are modified later, the legend labels may not correspond to the actual plots. 
# A better way is to add labels while plotting each curve.
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal') 
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

# In this code the legend labels are directly associated with each plot

ax.legend();
```

- Legend function can also take parameters like loc to specify the location of the legend on the plot.
- Eg: ax.legend(loc='upper left');
- Most common locations include: 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
- You can specify the location of the legend using the loc parameter in the legend() function.
    - common loc values are:
        - 'best'         0  ( Optimal location chosen by matplotlib )
        - 'upper right'  1
        - 'upper left'   2
        - 'lower left'   3
        - 'lower right'  4
        - 'right'        5
        - 'center left'  6
        - 'center right' 7
        - 'lower center' 8
        - 'upper center' 9
        - 'center'      10
- To adjust the location manually, you can use loc=(x,y) where x and y are coordinates between 0 and 1
- x represents horizontal position and y represents vertical position
- Eg: ax.legend(loc=(0.5, 0.5));  # places legend at center of the plot

```python
fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal') 
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')
ax.legend(loc=(0.5, 0.5));
```

Control colours
- Different colours can be used in matplotlib plots by specifying the colour parameter in the plot() function. 
- Colours can be specified using colour names, hex colour codes, or RGB tuples.
- Colour specifications can be called as colour name or abbreviation, hex colour code, or RGB tuple.
     1. Colour names or abbreviations: e.g., 'blue' or 'b' for blue colour. Some common colour names and their
        - blue - b
        - green - g
        - red - r
        - cyan - c
        - magenta - m
        - yellow - y
        - black - k
        - white - w
     2. Hex colour codes: '#RRGGBB' format, e.g., '#FF5733' for a shade of orange
     3. RGB tuples: (R, G, B) format where R, G, B are floats between 0 and 1, e.g., (0.1, 0.2, 0.5)

```python
x16 = np.arange(1, 5)

plt.plot(x16, 'r')
plt.plot(x16+1, 'g')
plt.plot(x16+2, 'b')

plt.show()
```

Control Line styles and Colors in Matplotlib Plots
- Line styles can be controlled using format strings in the plot() function
- Format string is a combination of color, marker and line style
- Example format strings: 'ro--' (red color, circle marker, dashed line), 'g^:' (green color, triangle marker, dotted line)

```python
x16 = np.arange(1, 5)
plt.plot(x16, 'r--', x16+1, 'g-.', x16+2, 'b:')    # red dashed, green dash-dot, blue dotted
plt.show()
```
